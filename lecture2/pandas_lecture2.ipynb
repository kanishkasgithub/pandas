### Filtering a DataFrame

```python
ipl.head(2)
```

```python
# find all the final winners
ipl[ipl['MatchNumber'] == 'Final'][['Season','WinningTeam']]
```

```python
# how many super over finishes have occured
ipl[ipl['SuperOver'] == 'Y'].shape[0]
```

```python
# how many matches has csk won in kolkata
ipl[(ipl['City'] == 'Kolkata') & (ipl['WinningTeam'] == 'Chennai Super Kings')].shape[0]
```

```python
# toss winner is match winner in percentage
(ipl[ipl['TossWinner'] == ipl['WinningTeam']].shape[0]/ipl.shape[0])*100
```

```python
# movies with rating higher than 8 and votes>10000
movies.head(2)
```

```python
movies[(movies['imdb_rating'] > 8) & (movies['imdb_votes'] > 10000)].shape[0]
```

```python
movies.head(1)
```

```python
# Action movies with rating higher than 7.5
action_movies = movies['genres'].str.split('|').apply(lambda x:'Action' in x)
movies[action_movies & (movies['imdb_rating'] > 7.5)].shape[0]
```

```python
movies['genres'].str    # explicitly converting into string
```

```python
action_movies = movies['genres'].str.contains('Action')
movies[action_movies & (movies['imdb_rating'] > 7.5)].shape[0]
```

```python
# write a function that can return the track record of 2 teams against each other
```

```python
def get_team_record(team1,team2):
  total_matches =  ipl[(ipl['Team1'] == team1) & (ipl['Team2'] == team2) | (ipl['Team1'] == team2) & (ipl['Team2'] == team1)]
  team1_won = total_matches[total_matches['WinningTeam'] == team1].shape[0]
  team2_won = total_matches[total_matches['WinningTeam'] == team2].shape[0]
  return f'Played {total_matches.shape[0]} matches against each other, {team1} won {team1_won} matches and {team2} won {team2_won} matches'
```

```python
team1 = 'Royal Challengers Bangalore' #'Gujarat Titans'  #'Chennai Super Kings'
team2 = 'Mumbai Indians'  #'Rajasthan Royals'
get_team_record(team1,team2)
```

### Important DataFrame Functions

**astype**

```python
# astype - to reduce memory footprint
ipl.info()
```

```python
ipl['ID'] = ipl['ID'].astype('int32')
```

```python
ipl.info()
```

Use category instead of object for columns having repeared values.

```python
ipl['Season'] = ipl['Season'].astype('category')
ipl['Team1'] = ipl['Team1'].astype('category')
ipl['Team2'] = ipl['Team2'].astype('category')
ipl['TossWinner'] = ipl['TossWinner'].astype('category')
ipl['WinningTeam'] = ipl['WinningTeam'].astype('category')
```
