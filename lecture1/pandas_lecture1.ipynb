### Pandas Series

A Pandas Series is like a column in a table. It is a 1-D array holding data of any type.

### Importing Pandas

````python
import numpy as np
import pandas as pd
````

### Series from lists

````python
# string
country = ['India','Pakistan','USA','Nepal','Srilanka']

pd.Series(country)
````

In Pandas, there is Series class which is used to create pandas series.

In Pandas series, there are two fields:

1.   Index - index for each value (by default generated automatically)
2.   Value - value which we pass

dtype: which gives datatype of value.

````python
# integers
runs = [13,24,56,78,100]
runs_series = pd.Series(runs)
runs_series
````

````python
# custom index
marks = [67,57,89,100]
subjects = ['maths','english','science','hindi']
pd.Series(marks,index=subjects)    # subjects as index
````

````python
# setting a name
marks = pd.Series(marks,index=subjects,name="Student's marks")
marks
````

### Series from dict

````python
marks = {'maths':67,'english':57,'science':89,'hindi':100}
marks_series = pd.Series(marks,name="Student's marks")
marks_series
````

### Copy and Views

````python
temp
````

````python
view_first_three = temp.head(3)
view_first_three
````

````python
view_first_three[0] = 100   # change view as well as original temp
````

````python
view_first_three
````

````python
temp
````

````python
copy_first_three = temp.head(3).copy()
copy_first_three
````

````python
copy_first_three[0] = 500   # change copy only, temp will not update
````

````python
copy_first_three
````

````python
temp
````

````python
s = pd.Series([1,2,3,4,5,6,7])
print(s)
v = s.head(5)
v[0] = 100
print(v)
print(s)
````

````python
c = s.head(5).copy()
c[0] = 200
print(c)
print(s)
````

View - refer to original series ( affect original series)

Copy - refer to copy / new series ( dont affect original series)
