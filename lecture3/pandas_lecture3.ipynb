```python
points_df = pd.DataFrame(
    {
        '1st point':[(3,4),(-6,5),(0,0),(-10,1),(4,5)],
        '2nd point':[(-3,4),(0,0),(2,2),(10,10),(1,1)]
    }
)

points_df
```

```python
def distance(row):
  point1 = row['1st point']
  point2 = row['2nd point']
  return ((point1[0]-point2[0])**2 + (point1[1]-point2[1])**2)**0.5
```

```python
points_df['distance'] = points_df.apply(distance,axis=1)
```

```python
points_df
```
