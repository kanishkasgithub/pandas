### Applying builtin aggregation fuctions on groupby objects

We can apply any aggregation functions such as sum, mean, min, max, count etc on groupby object. It will give dataframe which contains aggregated value of columns for each group.

```python
# Applying builtin aggregation fuctions on groupby objects - sum, mean, min, max, median, std, var, mode, count
genres.min()
```

```python
type(genres.min())
```

```python
genres.mean()
```

```python
genres.mean(numeric_only=True)
```

```python
# find the top 3 genres by total earning
genres.sum(numeric_only=True)['Gross'].sort_values(ascending=False).head(3)
```

```python
genres['Gross'].sum().sort_values(ascending=False).head(3)  # this is faster - use this approach in future
```

```python
# find the genre with highest avg IMDB rating
genres['IMDB_Rating'].mean().sort_values(ascending=False).head(1)
```

```python
# find director with most popularity
movies.groupby('Director')['No_of_Votes'].sum().sort_values(ascending=False).head(1).index[0]
```

```python
movies.groupby('Star1')['Series_Title'].count().sort_values(ascending=False)
```

### agg method

```python
genres.sum(numeric_only=True)
# apply sum on each numeric column, but suppose we want different aggregation operations on different columns -> use agg method
```

```python
# agg method
# passing dict
genres.agg({
    'Runtime':'mean',
    'IMDB_Rating':'mean',
    'No_of_Votes':'sum',
    'Gross':'sum',
    'Metascore':'max'
})
```

```python
# passing list of functions
genres.agg(['min','max','mean','sum'])
```

```python
# Adding both the syntax
genres.agg(
    {
        'Runtime':['min','mean'],
        'IMDB_Rating':'mean',
        'No_of_Votes':['sum','max'],
        'Gross':'sum',
        'Metascore':'min'
    }
)
```

### Looping on Groups

```python
# looping on groups
for group,data in genres:
  print(type(group),type(data))
```

When we loop groupby object, it will give two things:

1.   group - string eg. genre name
2.   data - which is dataframe contains all data of a group


```python
for group,data in genres:
  print(group)
```

```python
for group,data in genres:
  print(data)
```

```python
# find the highest rated movie of each genre
df = pd.DataFrame()
# df = pd.DataFrame(columns=movies.columns)
for group,data in genres:
  temp = data[data['IMDB_Rating'] == data['IMDB_Rating'].max()]
  df = pd.concat([df,temp])
df
```

```python
def highest_rated(group):
  return group[group['IMDB_Rating'] == group['IMDB_Rating'].max()]
```

```python
genres.apply(highest_rated)
```

### apply method

Split -> Apply -> Combine Strategy

```python
# split (apply) combine
# apply -> builtin function

genres.apply(min)
```

```python
genres.apply(np.minimum.reduce)
```

```python
# find number of movies starting with A for each group
def start_with_a(group):
  # return group[group['Series_Title'].str.startswith('A')]['Series_Title'].count()
  return group['Series_Title'].str.startswith('A').sum()
```

```python
genres.apply(start_with_a)
```

```python
movies[movies['Series_Title'].str.startswith('A')].groupby('Genre').size()
```

```python
movies[movies['Series_Title'].str.startswith('A')]['Genre'].value_counts()
```

```python
# find ranking of each movie in the group according to IMDB score
def genre_rank(group):
  group['genre_rank'] = group['IMDB_Rating'].rank(ascending=False)
  return group
```

```python
genres.apply(genre_rank)
```

```python
# find normalized IMDB rating group wise
def normal_rank(group):
  group['normalized_rating'] = (group['IMDB_Rating'] - group['IMDB_Rating'].min()) / (group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
  return group
```

```python
genres.apply(normal_rank)
```

```python
# Create a function that can return the highest score of any batsman
def highest_score(name):
  df_batter = ipl[ipl['batsman'] == name]
  return df_batter.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
```

```python
name = input('Enter batsman name: ')
highest_score(name)
```

```python
# find batsman record against all teams -> total runs, highest score, no of 4s, no of 6s
def record(group):
  group['total_runs']= group['batsman_runs'].sum()
  group['highest_score'] = group.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
  group['no_of_4'] = (group['batsman_runs']==4).sum()
  group['no_of_6'] = (group['batsman_runs']==6).sum()
  return group[['total_runs','highest_score','no_of_4','no_of_6']].drop_duplicates()
```

```python
df = ipl[ipl['batsman'] == 'MS Dhoni']
df.groupby('bowling_team').apply(record)
```

```python
res = pd.DataFrame()
for group,data in df.groupby('bowling_team'):
  total_runs = data['batsman_runs'].sum()
  highest_score= data.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
  no_of_4 = (data['batsman_runs']==4).sum()
  no_of_6 = (data['batsman_runs']==6).sum()
  temp = pd.DataFrame(data={'Total Runs':total_runs,'Highest Score':highest_score,'Total 4s':no_of_4,'Total 6s':no_of_6},index=[group])
  res = pd.concat([res,temp])
res
```
